### Predicting Temperature in London


A medida que el clima cambia, predecir el tiempo se vuelve cada vez más importante para las empresas. Dado que el tiempo depende de muchos factores diferentes, será necesario realizar numerosos experimentos para determinar cuál es el mejor enfoque para predecirlo. En este proyecto, realizará experimentos con diferentes modelos de regresión que predicen la temperatura media, utilizando una combinación de sklearn y MLflow.

Trabajará con los datos almacenados en london_weather.csv, que contiene las siguientes columnas:

- date: fecha de registro de la medición (int)
- cloud_cover: medición de la nubosidad en octas (float)
- sunshine: medición de la insolación en horas (hrs) (float)
- global_radiation: medición de la irradiancia en vatios por metro cuadrado (W/m²) (float)
- max_temp: temperatura máxima registrada en grados Celsius (°C) (float)
- mean_temp: temperatura media en grados Celsius (°C) (float)
- min_temp: temperatura mínima registrada en grados Celsius (°C) (float)
- precipitation: medición de la precipitación en milímetros (mm) (float)
- pressure: medición de la presión en pascales (Pa) (float)
- snow_depth: medición de la profundidad de la nieve en centímetros (cm) (float)

Utiliza el aprendizaje automático para predecir la temperatura media en Londres, Inglaterra, registrando tus métricas de error cuadrático medio (RMSE) con mlflow.

- Crea un modelo para predecir "mean_temp" con un RMSE de 3 o menos.
- Utiliza MLflow para registrar cualquier modelo que entrenes, sus hiperparámetros y sus respectivas puntuaciones RMSE (asegúrate de incluir "rmse" como parte del nombre de la métrica).
- Busca en todas tus ejecuciones de mlflow y almacena los resultados como una variable llamada experiment_results.

```python
# ============================================
# 📦 Importación de librerías necesarias
# ============================================
import pandas as pd                   # Manejo de datos tabulares (dataframes)
import numpy as np                    # Operaciones numéricas
import mlflow                         # Seguimiento de experimentos
import mlflow.sklearn                 # Módulo específico para modelos sklearn
import matplotlib.pyplot as plt        # Visualización de gráficos
import seaborn as sns                 # Visualización avanzada de datos
from sklearn.model_selection import train_test_split   # División de datos
from sklearn.metrics import root_mean_squared_error    # Métrica de error RMSE
from sklearn.impute import SimpleImputer               # Imputación de valores faltantes
from sklearn.preprocessing import StandardScaler       # Normalización (no se usa aquí)
from sklearn.linear_model import LinearRegression      # Modelo de regresión lineal
from sklearn.tree import DecisionTreeRegressor         # Árbol de decisión
from sklearn.ensemble import RandomForestRegressor     # Bosque aleatorio

# ============================================
# 📂 Carga del dataset
# ============================================
weather = pd.read_csv("../data/london_weather.csv")

# ============================================
# 🔍 Análisis exploratorio y limpieza de datos
# ============================================
weather.info()  # Verifica tipos de datos y valores nulos

# Convertir la columna 'date' a formato datetime (año-mes-día)
weather["date"] = pd.to_datetime(weather["date"], format="%Y%m%d")

# Extraer el año y el mes para análisis temporal
weather["year"] = weather["date"].dt.year
weather['month'] = weather['date'].dt.month

# ============================================
# 📊 Agregación de datos: calcular promedios mensuales
# ============================================
weather_metrics = [
    'cloud_cover', 'sunshine', 'global_radiation',
    'max_temp', 'mean_temp', 'min_temp',
    'precipitation', 'pressure', 'snow_depth'
 ]

# Agrupar por año y mes y calcular la media de cada métrica
weather_per_month = weather.groupby(['year', 'month'], as_index=False)[weather_metrics].mean()

# ============================================
# 🎨 Visualización de relaciones
# ============================================

# Gráfico de la temperatura media a lo largo de los años
plt.figure(figsize=(10, 6))  # Cambia el tamaño general del gráfico
sns.lineplot(x="year", y="mean_temp", data=weather_per_month, ci=None)

# Personalización de texto
plt.title("Temperatura media anual en Londres", fontsize=18, fontweight='bold')
plt.xlabel("Año", fontsize=10)
plt.ylabel("Temperatura media (°C)", fontsize=14)

# Cambiar tamaño de los ticks de los ejes
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.tight_layout()
plt.show()

# Mapa de calor de correlaciones entre variables
plt.figure(figsize=(10, 6))  # Cambia el tamaño general del gráfico
sns.heatmap(weather.corr(), annot=True)
# Cambiar tamaño de los ticks de los ejes
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.tight_layout()
plt.show()

# ============================================
# 🧩 Selección de características y variable objetivo
# ============================================
feature_selection = ['month', 'cloud_cover', 'sunshine', 'precipitation', 'pressure', 'global_radiation']
target_var = 'mean_temp'

# Eliminar filas donde 'mean_temp' esté vacía
weather = weather.dropna(subset=['mean_temp'])

# Definir los conjuntos de entrada (X) y salida (y)
X = weather[feature_selection]    
y = weather[target_var]

# ============================================
# ✂️ División de datos en entrenamiento y prueba
# ============================================
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=1
)

# ============================================
# 🧮 Imputación de valores faltantes
# ============================================
imputer = SimpleImputer(strategy="mean")   # Sustituye nulos con la media
X_train = imputer.fit_transform(X_train)
X_test  = imputer.transform(X_test)

# ============================================
# 🤖 Entrenamiento y registro de modelos con MLflow
# ============================================

# Se prueban tres profundidades distintas para los modelos basados en árboles
for idx, depth in enumerate([1, 2, 10]): 
    run_name = f"run_{idx}"  # Nombre del experimento

    # Inicia un nuevo "run" (ejecución) en MLflow
    with mlflow.start_run(run_name=run_name):
        
        # 1️⃣ Entrenar modelos
        lin_reg = LinearRegression().fit(X_train, y_train)
        tree_reg = DecisionTreeRegressor(random_state=42, max_depth=depth).fit(X_train, y_train)
        forest_reg = RandomForestRegressor(random_state=42, max_depth=depth).fit(X_train, y_train)
        
        # 2️⃣ Registrar modelos en MLflow
        mlflow.sklearn.log_model(lin_reg, name="lin_reg")
        mlflow.sklearn.log_model(tree_reg, name="tree_reg")
        mlflow.sklearn.log_model(forest_reg, name="forest_reg")
        
        # 3️⃣ Evaluar desempeño (RMSE)
        y_pred_lin_reg = lin_reg.predict(X_test)
        lin_reg_rmse = root_mean_squared_error(y_test, y_pred_lin_reg)
        
        y_pred_tree_reg = tree_reg.predict(X_test)
        tree_reg_rmse = root_mean_squared_error(y_test, y_pred_tree_reg)
        
        y_pred_forest_reg = forest_reg.predict(X_test)
        forest_reg_rmse = root_mean_squared_error(y_test, y_pred_forest_reg)
        
        # 4️⃣ Registrar parámetros y métricas en MLflow
        mlflow.log_param("max_depth", depth)  # Hiperparámetro
        mlflow.log_metric("rmse_lr", lin_reg_rmse)   # RMSE de regresión lineal
        mlflow.log_metric("rmse_tr", tree_reg_rmse)  # RMSE de árbol de decisión
        mlflow.log_metric("rmse_fr", forest_reg_rmse)  # RMSE de random forest

# ============================================
# 🔎 Consultar los resultados registrados
# ============================================
experiment_results = mlflow.search_runs()
experiment_results
```
